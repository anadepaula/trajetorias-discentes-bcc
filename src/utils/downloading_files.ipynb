```python
import sys
from pathlib import Path

repo_root = Path(".").absolute().parent.parent.as_posix()
sys.path.insert(0, repo_root)
```

```python
import os

from yaml import safe_load
from src.utils.file_downloader import download_file
from src.utils.logger import logger
```

```python
file_matriculas_cfes = repo_root / Path("matriculas_cfes.yaml")
with open(file_matriculas_cfes) as f:
    content = f.read()
    matriculas_cfes = safe_load(content)
```

```python
matriculas_cfes
```

```python
data_dir = "data"
dir_to_save = "matriculas_cfes"
path_to_save = Path(repo_root, data_dir, dir_to_save)
```

```python
col_edital = "edital"
col_resultado = "resultado"
```

```python
for ano, ano_data in matriculas_cfes.items():
    for quadri, quadri_data in ano_data.items():
        logger.info(f"downloading {ano} {quadri} {col_edital}")
        filename_edital = Path(path_to_save, f"{ano}{quadri}_{col_edital}.pdf")
        download_file(quadri_data[col_edital], filename_edital)

        logger.info(f"downloading {ano} {quadri} {col_resultado}")
        filename_resultado = Path(path_to_save, f"{ano}{quadri}_{col_resultado}.pdf")
        download_file(quadri_data[col_resultado], filename_resultado)
```
