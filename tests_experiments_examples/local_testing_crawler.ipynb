```python
soup = BeautifulSoup(
    response.text,
    'html.parser'
)
```

```python
paragraphs = soup.find_all("p")
```

```python
files = []
for p_id, paragraph in enumerate(paragraphs):
    children = paragraph.find_all()
    if len(children) == 1 and children[0].attrs.get("href") is not None and children[0].attrs.get("href").strip().endswith("csv"):
        files.append(children[0].attrs.get("href"))
```

```python
url_root = 'http://dados.ufabc.edu.br'
```

```python
paths = []
for file in files:
    if not file.startswith("http"):
        file = url_root + file
    paths.append(file)
```
