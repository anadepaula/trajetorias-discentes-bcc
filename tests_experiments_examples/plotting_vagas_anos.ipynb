```python
import sys
from pathlib import Path

repo_root = Path(".").absolute().as_posix()
sys.path.insert(0, repo_root)
```

```python
from yaml import safe_load
```

```python
from src.utils.logger import logger
```

```python
import pandas as pd
```

```python
file_matriculas_cfes = repo_root / Path("vagas_cfes.yaml")
with open(file_matriculas_cfes) as f:
    content = f.read()
    vagas_cfes = safe_load(content)
```

```python
entries = []
for ano, quadris_ano in vagas_cfes.items(): 
    # print(ano)
    for quadri, quadri_info in quadris_ano.items():
        data = dict(
            ano=ano,
            quadri=quadri,
            **quadri_info,
        )
        entries.append(data)
```

```python
df = pd.DataFrame(entries)
```
